```python
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from torch import optim
```

```python
iris = datasets.load_iris()
iris_data = iris.data
iris_target = iris.target
iris_labels = iris.target_names
```

**Neural Network Model**

```python
class NNIris(nn.Module):
  def __init__(self, input_layers_size, hidden1_layers_size, hidden2_layers_size, output_layers_size):
    super(NNIris, self).__init__()
    
    self.input_layer = nn.Linear(input_layers_size, hidden1_layers_size)
    self.hidden1_layers = nn.Linear(hidden1_layers_size, hidden2_layers_size)
    self.hidden2_layers = nn.Linear(hidden2_layers_size, output_layers_size)
    
    self.sigmoid1 = nn.Sigmoid()
    self.sigmoid2 = nn.Sigmoid()


  def forward(self, input):
    out = self.input_layer(input)
    out = self.sigmoid1(out)
    out = self.hidden1_layers(out)
    out = self.sigmoid2(out)
    out = self.hidden2_layers(out)
    return out


  def __normalize(self, data_set):
    __data = data_set.data
    __target = data_set.target
    __min_max_scaler = MinMaxScaler(feature_range=(0, 1))
    

    self.X_train, self.X_test, self.Y_train, self.Y_test = train_test_split(__data, __target)
    X_train_normalized_tensor = torch.tensor(__min_max_scaler.fit_transform(self.X_train), dtype=torch.float32)
    Y_train_normalized_tensor = torch.tensor(self.Y_train, dtype=torch.float32)

    return X_train_normalized_tensor, Y_train_normalized_tensor 

  def learn(self, epochs, data_set, lr, momentum):
    optimizer = optim.SGD(self.parameters(), lr=lr, momentum=momentum)
    criterion = nn.MSELoss()
    self.train()
    X_train_normalized, Y_train_normlized = self.__normalize(data_set=data_set)

    for epoch in range(epochs):
      first_loss = 0.0
      for i in range(0, X_train_normalized.size()[0]):
        optimizer.zero_grad()
        outputs = self.forward(X_train_normalized[i])
        loss = criterion(outputs, Y_train_normlized[i])
        loss.backward()
        optimizer.step()

        first_loss += loss

        if i  == X_train_normalized.size()[0] - 1:
          last_result = f"[{epoch+1}, {i+1}] - {first_loss/X_train_normalized.size()[0]}"
          print(last_result)
          first_loss = 0.0
      
    return last_result




    


```

```python
network = NNIris(4, 100, 50, 3)
network.learn(epochs=20000, data_set=iris, lr=0.01, momentum=0.5)
```
