```python
iris = datasets.load_iris()
iris_data = iris.data
iris_target = iris.target
iris_labels = iris.target_names
```

**Neural Network Model**

```python
class NNIris(nn.Module):
  def __init__(self, input_layers_size, hidden1_layers_size, hidden2_layers_size, output_layers_size):
    super(NNIris, self).__init__()
    
    self.input_layer = nn.Linear(input_layers_size, hidden1_layers_size)
    self.hidden1_layers = nn.Linear(hidden1_layers_size, hidden2_layers_size)
    self.hidden2_layers = nn.Linear(hidden2_layers_size, output_layers_size)
    
    self.sigmoid1 = nn.Sigmoid()
    self.sigmoid2 = nn.Sigmoid()

  def forward(self, input):
    out = self.input_layer(input)
    out = self.sigmoid1(out)
    out = self.hidden1_layers(out)
    out = self.sigmoid2(out)
    out = self.hidden2_layers(out)
    return out

  def normalize(self, data_set):
    self.data = data_set.data
    self.target = data_set.target

    self.X_train, self.X_test, self.Y_train, self.Y_test = train_test_split(self.data, self.target)

    return torch.tensor(self.X_train, dtype=torch.float32), torch.tensor(self.Y_train, dtype=torch.float32)

    


```

```python
network = NNIris(4, 100, 50, 3)
optimizer = optim.SGD(network.parameters(), lr=0.01, momentum=0.5)
criterion = nn.MSELoss()

data_train, labels_train = network.normalize(iris)

def fullPass(data_train, labels_train):
  first_loss = 0.0
  for i in range(0, data_train.size()[0]):
    optimizer.zero_grad()
    outputs = network.forward(data_train[i])
    loss = criterion(outputs, labels_train[i])
    loss.backward()
    optimizer.step()

    # print statistics
    first_loss += loss
    if i % data_train.size()[0] == data_train.size()[0]-1: 
        print(f'[{epoch+1}, {i+1}] - {first_loss/data_train.size()[0]}')
        first_loss = 0.0



network.train()
for epoch in range(100):
  fullPass(data_train, labels_train)
```
